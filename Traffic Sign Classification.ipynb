# 3. Image Preprocessing

### We have to apply shuffling as images are in order category wise

```python
from sklearn.utils import shuffle
X_train, y_train = shuffle(X_train,y_train)
```

### Convert Images into grey scale from RGB format 

##### For White Color
- 255,255,255(RGB) = White color = 255(Grey Scale)
- formula sum(RGB colors)/3 => (255+255+255)/3 = 255(White color in grey scale) 

##### For black Color
- 0,0,0, (RGB) = Black Color = 0(grey Scale)
- (0+0+0)/3 = 0 = 0(Black in grey scale)


```python
X_train[0][0][0]
# RGB Format
```

```python
print(X_train.shape)
X_train_grey = np.sum(X_train/3, axis=3, keepdims=True)
print(X_train_grey.shape)
```

```python
X_train_grey[0][0][0]
# Grey Scale
```

```python
X_test_grey = np.sum(X_test/3, axis=3, keepdims=True)
X_valid_grey = np.sum(X_valid/3, axis=3, keepdims=True)
```

### Normalization

#### Convert Image into -1 to 1 range
- If we want range from 0 to 1, we can divide by 255. 
- If we want range between -1 to 1, then do the half of 255/2 = 127.5 ~ 128.0. Now to convert my data into -1 to 1 range, this formula will be useful (number-128)/128. 


- Ex:-> 0 -> (0-128)/128 = -1 -> Lowest is 0 and it became -1.

      128 -> (128-128)/128 = 0 -> Middle value 128 will become 0.

      255 -> (255-128)/128 = 0.99 ~ 1 -> Highest value 255 became 1.



```python
half_num = round(X_train_grey.max()/2)
X_train_grey_norm = (X_train_grey-half_num)/half_num

half_num = round(X_valid_grey.max()/2)
X_valid_grey_norm = (X_valid_grey-half_num)/half_num

half_num = round(X_test_grey.max()/2)
X_test_grey_norm = (X_test_grey-half_num)/half_num

print("%.2f"%(X_train_grey_norm.min()),"\t","%.2f"%(X_valid_grey_norm.min()),"\t","%.2f"%(X_test_grey_norm.min()))
print("%.2f"%(X_train_grey_norm.max()),"\t","%.2f"%(X_valid_grey_norm.max()),"\t","%.2f"%(X_test_grey_norm.max()))
```

# 4. Model Training

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, AveragePooling2D, Dense, Flatten, Dropout # Regularization technique-> To prevent overfitting (Removes Neurons) 
from keras.optimizers import Adam
from keras.callbacks import TensorBoard
```

```python
classifier = Sequential()

# Convolution Layer and Pooling Layer
classifier.add(Conv2D(filters=6, kernel_size=(5,5), activation='relu', input_shape=(32,32,1)))
classifier.add(AveragePooling2D())

# Second Convolution and Pooling Layer
classifier.add(Conv2D(filters=16, kernel_size=(5,5), activation='relu'))
classifier.add(AveragePooling2D())

# Flattening
classifier.add(Flatten())

# Hidden Layers
classifier.add(Dense(units=120, activation='relu'))
classifier.add(Dense(units=84, activation='relu'))

# Output Layer
classifier.add(Dense(units=43, activation='softmax'))
```

#### If your targets are one-hot encoded, use categorical_crossentropy.
- Examples of one-hot encodings:
[1,0,0]
[0,1,0]
[0,0,1]

#### But if your targets are integers, use sparse_categorical_crossentropy.
- Examples of integer encodings (for the sake of completion):
1
2
3

```python
classifier.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

### Training the Model

```python
history = classifier.fit(X_train_grey_norm, 
               y_train, 
               batch_size= 500, 
               epochs=50, 
               verbose=1,  #verbose: Integer. 0, 1, or 2. Verbosity mode. 0 = silent, 1 = progress bar, 2 = one line per epoch.
               validation_data=(X_valid_grey_norm, y_valid))
```

# 5. Evaluation of the model

```python
score = classifier.evaluate(X_test_grey_norm, y_test)
print(score)
print("Accuracy : ",score[1])
```

#### Predict Classes 

```python
y_pred = classifier.predict_classes(X_test_grey_norm)
```

#### Confusion Matrix Evaluation

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
TP = 0
for i,row in enumerate(cm):
    for j,col in enumerate(row):
        if i==j:
            TP+=col
TP
```

```python
cm_accuracy = (100*TP)/len(y_pred)
cm_accuracy
```
